```python
data=data.rename({'R&D Spend':'RDS','Administration':'Admin','Marketing Spend':'MS','Profit':'PT'},axis = 1)
data.head()
```

# TRANSFORMATION

```python
#since state is categorical coverting to numerical using dummification
data2=pd.get_dummies(data,columns=['State'],drop_first = True)
data2.head()
```

```python
#check corr with new dataset
data2.corr()
```

# Building our first model

```python
model_1=smf.ols('PT~RDS+Admin+MS',data=data).fit()
```

```python
model_1.params
```

```python
model_1.summary()
```

# Computing VIF

```python
#calculating vif values for independent variables
#collinearity check
rsq_RDS = smf.ols("RDS~Admin+MS",data = data2).fit().rsquared
vif_RDS = 1/(1-rsq_RDS)
```

```python
rsq_Admin = smf.ols('Admin~RDS+MS',data=data2).fit().rsquared
vif_Admin = 1/(1-rsq_RDS)
```

```python
rsq_MS = smf.ols('MS~Admin+RDS',data=data2).fit().rsquared
vif_MS = 1/(1-rsq_MS)
```

```python
#sorting VIF values in a DataFrame
d1 = {"Features":['RDS','Admin','MS'],'VIF':[vif_RDS,vif_Admin,vif_MS]}
vif=pd.DataFrame(d1)
vif
```

- since none of the features have vif>10 we have to consider all Features

```python
#checking for normality of residuals by qq plot
sm.qqplot(model_1.resid,line = 'q')
plt.title("Normal Q-Q plot of residuals")
```

```python
#test to check wheather it shows homoscedacity or heteroscedacity
def standard_values(vals): return(vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model_1.fittedvalues),standard_values(model_1.resid))
plt.title('Residual plot')
plt.xlabel('standard fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

# Model deletion diagnostics

```python
(c,_)=model_1.get_influence().cooks_distance
c
```

```python
#plot the influencers using the stem plot
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data2)),np.round(c,3))
plt.xlabel('row index')
plt.ylabel('cooks distance')
plt.show()
```

```python
#locating the value that influencing our plot
np.argmax(c),np.max(c)
```

```python
influence_plot(model_1)
plt.show
```

```python
# leverage cutoff value is 3(k+1)/n
k=data2.shape[1]
n=data2.shape[0]
leverage_cutoff=3*((k+1)/n)
leverage_cutoff
```

```python
#locating the influencer in our dataset
data2[data2.index.isin([48,49])]
```

```python
#Dropping the influencer and reset the dataset
data3= data2.drop(data.index[[48,49]],axis = 0).reset_index(drop=True)
data3
```

```python
#building our new model with our updated dataset
model_2=smf.ols('PT~RDS+Admin+MS',data=data3).fit()
```

```python
model_2.summary()
```

```python
(d,_)=model_2.get_influence().cooks_distance
d
```

```python
#plot the influencers using the stem plot 
fig = plt.figure(figsize = (10,7))
plt.stem(np.arange(len(data3)),np.round(d,3))
plt.xlabel('row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(d),np.max(d)
```
